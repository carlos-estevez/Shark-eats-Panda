**Regex : Sustituir valores de edades conflictivos**

```python
#Teen,teen,young
df.Age = df.Age.str.replace(r"\w{4,}", "Teenager")
```

```python
#Espacio
for x in df.Age:
    df.Age = df.Age.str.replace(r"\s", x[0][1:])
df.Age.value_counts()
```

```python
# 30s, 50s ...
for x in df.Age:
    df.Age = df.Age.str.replace(r"\d{2}s", x[0][:2])
```

```python
# age "or" age
for x in df.Age:
    df.Age = df.Age.str.replace(r"(\d{2}\s) or (\d{2})", x[0][:2])
df.Age.value_counts()
```

```python
#Groups:Children (:12), teenager(13-19), young adult (20-39), adult (40,64), elderly (65:)
def groupage(x):
    for i in x:
        try:
            if i < 13:
                return "Children"
            elif 13 >= i < 20: 
                return "Teenager"
            elif 20 >= i < 40: 
                return "Young adult"
            elif 40 >= i < 65:  
                return "Adult"
            elif i >=65:
                return "Elderly"
        except:
            pass
```

```python
df["Age"]=groupage(df["Age"])
df["Age"].value_counts()
```
